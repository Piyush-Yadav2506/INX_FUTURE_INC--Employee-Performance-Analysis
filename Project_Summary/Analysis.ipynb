# Analysis – Interpreting Data and Summary of Various Choices

## 1. ML Algorithm and Training Methods Used

- Multiple machine learning algorithms were evaluated, including **Logistic Regression, Decision Tree, Random Forest, Gradient Boosting, AdaBoost, SVM, and XGBoost**.
- The **best-performing model was XGBoost**, selected based on highest test accuracy, precision, and F1 score.

## 2. Most Important Features Selected & Why

### Top 3 Features Affecting Employee Performance:

1. **EmpLastSalaryHikePercent** – High salary hikes correlated with strong performance, indicating recognition and motivation are drivers.
2. **EmpJobSatisfaction & EmpWorkLifeBalance** – Both strongly correlated with better performance; low scores signaled higher underperformance and disengagement.
3. **YearsSinceLastPromotion** – Longer intervals without promotion correlated with underperformance, indicating career stagnation.

### Feature Engineering & Selection Techniques:

- Included data cleaning, handling missing values, fixing data types, and removing outliers.
- Transformations used: **Label Encoding, One-Hot Encoding, Standard Scaling**.
- Transformations done using **Column Transformer**

## 3. Other Techniques and Tools Used

- **Automated EDA**: Utilized **Sweetviz** for quick, visual insights into the data distributions and outliers.
- **Quality Checks**: Data processed to remove or clarify real outlier issues.
- **Business Problem Framing**: Segmented analysis to address department-wise performance, identify non-performing employees, and rank important performance factors.
- Used **joblib** for saving/loading models and encoders.

## 4. Feature Engineering Highlights

- Data was split into train/test using sklearn, ensuring robust validation.

- Final cleaned dataset saved for reproducibility.

## 5. Results, Analysis, and Insights

### Model Comparison:
- XGBoost and Gradient Boosting are the only models that consistently handle class 4 well.
- AdaBoost, SVM, and Logistic Regression struggle significantly with class 4 — low recall and F1.
- Gradient Boosting is a close second, slightly less robust on class 4.
- Random Forest is strong but slightly weaker on minority class.
- Decision Tree generalizes decently but overfits hard.
- Logistic Regression and SVM are interpretable but underperform.
- AdaBoost is not viable for use case as performance is poor.
- **Final Verdict XGBoost** is your **best model** — highest test F1, strong precision, and solid handling of all classes.
- **XGBoost outperformed** after hyperparameter tuning other models on test accuracy (93%) and generalization.

- Work-life balance is critical for retention and performance.
- Lack of promotion is a warning sign for HR intervention.
